## Unit Converter

```python
import math

#Radian to Degree
def rad_deg():
    rad= float(input("Enter the radian value:"))
    return rad*180*7/22
    
#Degrees to radian
def deg_rad():
    deg=float(input("Enter the degrees value"))
    return deg*22/(7*180)
    
#Number to its Factorial
def num_fact():
    num=int(input("Enter the number to find factorial:"))
    return (math.factorial(num))

#Number to its square root
def num_sqrt():
    num=int(input("Enter the number to convert to its square root."))
    return(math.sqrt(num))  

#Megabyte to nibble
def Meg_nib ():
    Mega =int(input("Enter the value in Megabyte"))
    return Mega*2000000
        
#Fraction to decimal 
def Frac_dec():
    Frac =input("Fraction value : ")
    return eval(Frac)
       
#Km/h to m/s
def Kmh_ms():
    Kmh= eval(input("Enter the value in Km/h "))
    return 1000/3600*Kmh

#defm/s to km/h
def ms_kmh():
    ms = eval(input("Enter the value in m/s "))
    return 3600/1000*ms 

#def int to float
def int_flo():
    val1 =int(input("Enter the integer value"))
    return float(val1)
    
#float to int 
def flo_int(): 
    val= float(input("Enter the float value"))
    return int(round(val))
    
#Decimal to Octal
def dec_oct():
    decimal = int(input("Enter any decimal number"))
    i = 1
    sum=0
    while decimal !=0:
        rem =int(decimal%8)
        sum = sum +i*rem
        i= i*10
        decimal = decimal/8
    return(f"Octal Number = {sum}")

#Octal to decimal    
def oct_deci():
   a=input('Enter the value in octadecimal')
   b=[]
   k=1
   for i in a:
      x=int(i)(8*(len(str(a))-k))
      k+=1
      b.append(x)
   return (f'the decemal number is {x}')

#decimal to hexadecimal     
def deci_hex():
   num=int(input("Enter the value in decimal"))
   return (hex(num))

#Hexadecimal to decimal
def hex_deci():
    hex = input("Enter the Hexadecimal number:").upper()
    l = len(hex)
    decimal = 0
    pos = 0
    for i in range(l - 1, -1, -1):
        if '0' <= hex[i] <= '9':
            digit = int(hex[i])
            decimal += digit * pow(16, pos)
            pos += 1
        elif 'A' <= hex[i] <= 'F':
            digit = ord(hex[i]) - 55
            decimal += digit * pow(16, pos)
            pos += 1
    return decimal
#decimal to binary 
def deci_bin():
   a=int(input("Enter the decimal number :"))
   return (bin(a))

#Binary to decimal
def bin_deci():
   a=input('Enter the binary number')
   b=[]
   k=1
   for i in a:
      x=int(i)(2*(len(str(a))-k))
      k+=1
      b.append(x)
   return (f'the decimal number is {x}')
   
#Decimal to ascII
def deci_ascII():
   a=int(input("Enter the decimal"))
   return print(chr(a))

#ascII to decimal 
def ascII_deci():
   a=input("Enter the ascII character")
   return (ord(a))

#Celsius to fahrenheit 
def cel_far():
   a=int(input("Enter the celsius"))
   b=(a*(9/5))+32
   return (f'{a} celsius is equal to {b} fahrenheit')

#Fahrenheit to celsius
def far_cel():
   a=int(input("Enter fahrenheit value"))
   b=(a-32)*(5/9)
   return (f'{a} fahrenheit is equal to {b} celsius')


dictionary={'1':rad_deg,'2': deg_rad,'3':num_fact,'4':num_sqrt,'5':Meg_nib,
          '6':Frac_dec,'7':Kmh_ms,'8':ms_kmh,'9':int_flo,'10':flo_int,
          '11':dec_oct,'12':oct_deci,'13':deci_hex,'14':hex_deci,'15':deci_bin,'16':bin_deci,
          '17':deci_ascII,'18':ascII_deci,'19':cel_far,'20':far_cel}
print('''welcome to the converter:
1 Radian to degree\n2 Degree to Radian\n3 Number to factorial\n4 Number to its square root\n5 Megabyte to nibble'
6 Fraction to decimal\n7 Kilometer per hour to meter per second \n8 meter per second to Kilometer per hour'
9 Integer to float\n10 float to integer\n11 Decimal to octal\n12 Octal to decimal\n13 decimal to hexadeciamal'
14 hexadeciamal to decimal\n15 decimal to Binary\n16 binary to decimal\n17 character to ASCII'
18 ASCII to character \n19 celsius to fahernheit\n20 fahernheit to celsius''')

converter=input(" Enter the function value to converter")
for i in dictionary:
   if converter==i:
       print(dictionary[i]())
   else :
        pass
       

       
```
